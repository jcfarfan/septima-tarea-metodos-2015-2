```python
n = 96
x = zeros(n)
y = zeros(n)
t = zeros(n)

t[0] = 0.006
t[95] = 0.799
x[0] = 15.0
y[0] = 13.0
h = (t[95]-t[0])/n
print h
```

```python
g = 0.1
alpha = g
beta = g
gamma = g
epsilon = g

def xprime(a,b):
    xp = a*(alpha-beta*b)
    return xp
def yprime(a,b):
    yp = -b*(gamma-epsilon*a) 
    return yp
```

```python
for i in range(1,n):
    k1 = xprime(x[i-1],y[i-1])
    l1 = yprime(x[i-1],y[i-1])
    k2 = xprime(x[i-1]+0.5*h*k1,y[i-1]+0.5*h*l1)
    l2 = yprime(x[i-1]+0.5*h*k1,y[i-1]+0.5*h*l1)
    k3 = xprime(x[i-1]+0.5*h*k2,y[i-1]+0.5*h*l2)
    l3 = yprime(x[i-1]+0.5*h*k2,y[i-1]+0.5*h*l2)
    k4 = xprime(x[i-1]+h*k3,y[i-1]+h*l3)
    l4 = yprime(x[i-1]+h*k3,y[i-1]+h*l3)
    
    x[i] = x[i-1] + (1.0/6.0)*(k1+2*k2+2*k3+k4)
    y[i] = y[i-1] + (1.0/6.0)*(l1+2*l2+2*l3+l4)
    t[i] = t[i-1] + h
print x
```

```python
scatter(x,y)
```
